#### Plotting: Compare the average Mean-F1 and Accuracy (Acc) of WCSGNet and WCSGNet(Logarithmic Transformation)"

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
plt.rcParams['font.family'] = 'Times New Roman'
save_folder = "../Results_wcsgnet"
f1_path = os.path.join(save_folder, "wcsgnet_F1_Acc", "Mean_F1_Scores.csv")
f1_results = pd.read_csv(f1_path, index_col=0)
dataset_names = ['Zhang T', 'Kang', 'Zheng 68k', 'Baron Human', "Muraro", 'AMB', 'TM', 'Baron Mouse']
method_names = ["WCSGNet", "WCSGNet(Logarithmic Transformation)"]

deep_f1_results = f1_results.loc[method_names]

WCSGNet = deep_f1_results.loc["WCSGNet", dataset_names]
WCSGNet_LT = deep_f1_results.loc["WCSGNet(Logarithmic Transformation)", dataset_names]

x = np.arange(len(dataset_names)) 
width = 0.35 

plt.figure(figsize=(6, 4))

plt.bar(x - width/2, WCSGNet, width, label='WCSGNet', color='#72B6A1', edgecolor='white', linewidth=1)
plt.bar(x + width/2, WCSGNet_LT, width, label='WCSGNet(Logarithmic Transformation)', color='#E99675', edgecolor='white', linewidth=1)
plt.xticks(x, dataset_names, fontsize=10, rotation=45)  
plt.ylim(0.6, 1.0)  

ax = plt.gca()
ax.spines['top'].set_visible(False)  
ax.spines['right'].set_visible(False)  

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.3), fontsize=10, ncol=2, frameon=False)
def add_value_labels(ax):
    for bar in ax.patches:
        height = bar.get_height()
        ax.annotate(f'{height:.3f}',  
                    xy=(bar.get_x() + bar.get_width() / 2, height),  
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8,  
                    rotation=90)  

add_value_labels(ax)
plt.tight_layout()

plt.savefig('../../result/Figures/LT_compare/LT_F1.png', format='png', dpi=1200, bbox_inches='tight')
plt.savefig('../../result/Figures/LT_compare/LT_F1.svg', format='svg', dpi=1200, bbox_inches='tight')

plt.show()
```
